```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

### Binomial Distribution
---

The p.m.f. of the binomial distribution is

\begin{equation*}
 f(x|n,p) = \binom{n}{x}p^x(1-p)^{n-x},\quad x=0,1,\dots,n,
\end{equation*}

where $0<p<1$.

```python
fig1 = plt.figure(num=1, facecolor='w')
bin_width = 0.25
colors = ['b', 'g', 'r']
n = 10
x = np.linspace(0, n, n + 1)
for idx, p in enumerate([0.5, 0.2, 0.8]):
    plt.bar(x + bin_width*(idx - 1.0), st.binom.pmf(x, n, p), width=bin_width, color=colors[idx],label='p = {0:<3.1f}'.format(p))
plt.xlim((-0.5, n + 0.5))
plt.ylim((0, 0.34))
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend(loc='upper center', frameon=False)
# plt.savefig('ms_fig_binom_pmf.eps', dpi=600)
plt.show()
```

### Negative Binomial Distribution
---

The p.m.f. of a negative binomial distribution is

\begin{equation*}
f(x|p) = \binom{x+r-1}{r-1}p^r(1-p)^x,\quad x=0,1,2,\dots
\end{equation*}

where $0 < p < 1$.


```python
fig2 = plt.figure(num=2, facecolor='w')
bin_width = 0.25
colors = ['b', 'g', 'r']
x = np.linspace(0, 10, 11)
for idx, r in enumerate([1.0, 3.0, 5.0]):
    plt.bar(x + bin_width*(idx - 1.0), st.nbinom.pmf(x, r, 0.5), width=bin_width, color=colors[idx], label='r = {0:<3.1f}'.format(r))
plt.xlim((-0.5, 10.5))
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend(loc='best', frameon=False)
# plt.savefig('ms_fig_nbin_pmf.eps', dpi=600)
plt.show()
```

### Poisson Distribution
---

The p.m.f. of a Poisson distribution is

\begin{equation*}
 f(x|\lambda) = \frac{e^{-\lambda}\lambda^x}{x!},\quad x = 0,1,2,\dots
\end{equation*}

where $\lambda>0$.


```python
fig3 = plt.figure(num=3, facecolor='w')
bin_width = 0.25
colors = ['b', 'g', 'r']
x = np.linspace(0, 12, 13)
for idx, lam in enumerate([1.0, 3.0, 5.0]):
    plt.bar(x + bin_width*(idx - 1.0), st.poisson.pmf(x, lam), width=bin_width, color=colors[idx], label='$\\lambda$ = {0:<3.1f}'.format(lam))
plt.xlim((-0.5, 12.5))
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend(loc='best', frameon=False)
# plt.savefig('ms_fig_poisson_pmf.eps', dpi=600)
plt.show()
```

### Uniform Distribution
---

The p.d.f. of a (continuous) uniform distribution on the interval $[a,b]$ is

\begin{equation*}
 f(x|a,b) = 
 \begin{cases}
 \displaystyle
 \frac1{b-a} & \text{for } a\leqq x\leqq b; \\
 0 & \text{otherwise.}
 \end{cases}
\end{equation*}


```python
fig4 = plt.figure(num=4, facecolor='w')
colors = ['b', 'g', 'r']
a = np.array([0.0, -1.5, -2.0])
b = np.array([1.0, -0.5, 2.0])
for idx in range(a.size):
    a_i = a[idx]
    b_i = b[idx]
    x_i = np.linspace(a_i, b_i, 11)
    plt.plot(x_i, st.uniform.pdf(x_i, loc=a_i, scale=b_i-a_i), color=colors[idx], label='(a,b) = ({0:<3.1f},{1:<3.1f})'.format(a_i,b_i))
    plt.plot([a_i, a_i], [0.0, 1.0/(b_i-a_i)], linestyle=':', color=colors[idx])
    plt.plot([b_i, b_i], [0.0, 1.0/(b_i-a_i)], linestyle=':', color=colors[idx])
plt.xlim((-2.5, 2.5))
plt.ylim((0.0, 1.5))
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend(loc='best', frameon=False)
# plt.savefig('ms_fig_uniform_pdf.eps', dpi=600)
plt.show()
```

### Exponential Distribution
---

The p.d.f. of an exponential distribution with parameter $\theta$ $(\theta>0)$ is

\begin{equation*}
 f(x|\theta) = \frac1{\theta} e^{-\frac{x}{\theta}},\quad  x > 0.
\end{equation*}


```python
fig5 = plt.figure(num=5, facecolor='w')
colors = ['b', 'g', 'r']
x = np.linspace(0, 3, 101)
for idx, the in enumerate([1.0, 0.5, 2.0]):
    plt.plot(x, st.expon.pdf(x, scale=the), color=colors[idx], label='$\\theta$ = {0:<3.1f}'.format(the))
plt.xlim((0.0, 3.0))
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend(loc='best', frameon=False)
# plt.savefig('ms_fig_exponential_pdf.eps', dpi=600)
plt.show()
```

### Normal Distribution
---

The p.d.f. of a normal (Gaussian) distribution is

\begin{equation*}
 f(x|\mu,\sigma^2)
 = \frac1{\sqrt{2\pi\sigma^2}}
 \exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right].
\end{equation*}


```python
fig6, ax6 = plt.subplots(2, 1, sharex='col', num=6, figsize=(8, 6), facecolor='w')
colors = ['b', 'g', 'r']
x = np.linspace(-6, 6, 201)
ax6[0].set_xlim((-6, 6))
for idx, mu in enumerate([0.0, 2.0, -2.0]):
    ax6[0].plot(x, st.norm.pdf(x, loc=mu), color=colors[idx], label='$\\mu$ = {0:< 3.1f}'.format(mu))
ax6[0].set_ylim((0, 0.45))
ax6[0].set_ylabel('Probability density')
ax6[0].legend(loc='upper right', frameon=False)
for idx, sigma in enumerate([1.0, 0.5, 2.0]):
    ax6[1].plot(x, st.norm.pdf(x, scale=sigma), color=colors[idx], label='$\\sigma$ = {0:<3.1f}'.format(sigma))
ax6[1].set_ylim((0, 0.85))
ax6[1].set_xlabel('x')
ax6[1].set_ylabel('Probability density')
ax6[1].legend(loc='upper right', frameon=False)
plt.tight_layout()
# plt.savefig('ms_fig_normal_pdf.eps', dpi=600)
plt.show()
```

+ $\mu$ is called the *location parameter*.
+ $\sigma$ is called the *scale parameter*.

### Normal Approximation of the Binomial Distribution
---

The binomial distribution converges to the normal distribution as $n$ goes to infinity.


```python
fig7, ax7 = plt.subplots(4, 1, sharex='col', num=7, figsize=(8, 6), facecolor='w')
ax7[0].set_xlim((-0.5, 20.5))
ax7[0].set_xticks(np.linspace(0, 20, 11))
for idx, n in enumerate([10, 20, 50, 100]):
    x = np.linspace(0, n, n + 1)
    ax7[idx].bar(x, st.binom.pmf(x, n, 0.1), color='b', label='n = {0:< 3.0f}'.format(n))
    ax7[idx].set_ylabel('Probability')
    ax7[idx].legend(loc='upper right', frameon=False)
ax7[-1].set_xlabel('x')
plt.tight_layout()
# plt.savefig('ms_fig_binom2normal.eps', dpi=600)
plt.show()
```

### Cumulative Distribution Function
---

1. Uniform distribution
\begin{equation*}
 F(x) = \begin{cases}
  0, & (x < a); \\
  \frac{x-a}{b-a}, & (a \leqq x \leqq b); \\
  1, & (x > b). \\
 \end{cases}
\end{equation*}

2. Exponential distribution
\begin{equation*}
 F(x) = \begin{cases}
 0, & (x < 0); \\
 1-e^{-x/\theta}, & (x\geqq 0).
 \end{cases}
\end{equation*}

3. Normal distribution
\begin{equation*}
 F(x) = \int_{-\infty}^x\frac1{\sqrt{2\pi\sigma^2}}\exp\left[-\frac{(t-\mu)^2}{2\sigma^2}\right]dt.
\end{equation*}

```python
fig8, ax8 = plt.subplots(3, 2, sharex='all', sharey='all', num=8, figsize=(8, 6), facecolor='w')
x1 = np.linspace(0.0, 4.0, 101)
x2 = np.linspace(-4.0, 4.0, 201)
ax8[0, 0].set_xlim((-4.0, 4.0))
ax8[0, 0].set_ylim((0.0, 1.1))
ax8[0, 0].plot([-4.0, 0.0], [0.0, 0.0], 'b-')
ax8[0, 0].plot([0.0, 1.0], [1.0, 1.0], 'b-')
ax8[0, 0].plot([1.0, 4.0], [0.0, 0.0], 'b-')
ax8[0, 0].plot([0.0, 0.0], [0.0, 1.0], 'b:')
ax8[0, 0].plot([1.0, 1.0], [0.0, 1.0], 'b:')
ax8[0, 0].set_title('Uniform p.d.f.')
ax8[0, 1].plot([-4.0, 0.0], [0.0, 0.0], 'b-')
ax8[0, 1].plot([0.0, 1.0], [0.0, 1.0], 'b-')
ax8[0, 1].plot([1.0, 4.0], [1.0, 1.0], 'b-')
ax8[0, 1].set_title('Uniform c.d.f.')
ax8[1, 0].plot([-4.0, 0.0], [0.0, 0.0], 'b-')
ax8[1, 0].plot([0.0, 0.0], [0.0, 1.0], 'b:')
ax8[1, 0].plot(x1, st.expon.pdf(x1), 'b-')
ax8[1, 0].set_title('Exponential p.d.f.')
ax8[1, 1].plot([-4.0, 0.0], [0.0, 0.0], 'b-')
ax8[1, 1].plot(x1, st.expon.cdf(x1), 'b-')
ax8[1, 1].set_title('Exponential c.d.f.')
ax8[2, 0].plot(x2, st.norm.pdf(x2), 'b-')
ax8[2, 0].set_xlabel('x')
ax8[2, 0].set_title('Normal p.d.f.')
ax8[2, 1].plot(x2, st.norm.cdf(x2), 'b-')
ax8[2, 1].set_title('Normal c.d.f.')
ax8[2, 1].set_xlabel('x')
plt.tight_layout()
# plt.savefig('ms_fig_pdf_vs_cdf.eps', dpi=600)
plt.show()
```
